CREATING BAR GRAPHS

```python
bb["Close"]
```

```python
def ranking(sp):
    if sp<=10:
        return "Poor"
    elif sp<=50:
        return "Satisfactory"
    else:
        return "Awesome"
```

```python
bb["Close"].apply(ranking).value_counts().plot()
```

```python
bb["Close"].apply(ranking).value_counts().plot(kind = "bar")
```

```python
bb["Close"].apply(ranking).value_counts().plot(kind = "barh")
```

CREATING PIE-CHARTS

```python
plt.style.use('dark_background')
bb["Close"].apply(ranking).value_counts().plot(kind = "pie")
```

```python
plt.style.use('classic')
bb["Close"].apply(ranking).value_counts().plot(kind = "pie", legend = True)
```
